Comando describe()

Permite avanzar sobre el conocimiento del dataset, para empezar a reconocer su distribución, trabajando solamente con las columnas/features/rasgos del tipo texto/string.






Definición de Percentiles:

El percentil es una medida de posición usada en estadística que indica, una vez ordenados los datos de menor a mayor, el valor de la variable por debajo del cual se encuentra un porcentaje dado de observaciones en un grupo.

Ejemplo: el percentil 20º es el valor bajo el cual se encuentran el 20% de las observaciones y el 80% restante son mayores.



Aparecen citados en la literatura científica por primera vez por Francis Galton en 1885.

     P25 = Q1

     P50 = Q2 = mediana

     P75 = Q3






Referencia Wikipedia - Percentil



```python
df[['Quantity','UnitPrice']].describe(percentiles=[0.05, 0.25, 0.50, 0.75, 0.95])
```

```python
df.describe(datetime_is_numeric=True, exclude=['object','float', 'int'], percentiles=[0.05, 0.25, 0.50, 0.75, 0.95])
```

```python
df.describe(include=['object'], exclude=['float', 'int'])
```
